## QUESTIONS | ANALYSIS

#### 1.What is the overall trend of malaria incidence across African countries over time

```python
# Convert the column to floats, handling non-numeric values by coercing them to NaN
clean_df['Incidence of malaria (per 1,000 population at risk)'] = \
pd.to_numeric(clean_df['Incidence of malaria (per 1,000 population at risk)'], errors='coerce')
```

```python
# Group by 'Year' and calculate the mean incidence of malaria
malaria_trend = clean_df.groupby('Year')['Incidence of malaria (per 1,000 population at risk)'].mean()
print(f"Overall trend of malaria incidence: {malaria_trend}")
```

#### 2.Which African country reported the highest number of malaria cases in a given year?

```python
# Convert the 'Malaria cases reported' column to floats, handling non-numeric values
clean_df['Malaria cases reported'] = pd.to_numeric(clean_df['Malaria cases reported'], errors='coerce')
```

```python
max_malaria_cases = clean_df[clean_df['Malaria cases reported'] == clean_df['Malaria cases reported'].max()]
max_malaria_cases
```

#### 3.What are the top 5 African countries with the highest Malaria Incidence?

```python
clean_df["Incidence of malaria (per 1,000 population at risk)"] = \
pd.to_numeric(clean_df["Incidence of malaria (per 1,000 population at risk)"], errors='coerce')
```

```python
country_incidence = clean_df.groupby("Country Name")["Incidence of malaria (per 1,000 population at risk)"].mean()
top_5_countries = country_incidence.nlargest(5)
top_5_countries
```

```python
clean_df
```

#### 4.Which year had the highest global average malaria incidence?

```python
highest_incidence_year = clean_df.groupby('Year')['Incidence of malaria (per 1,000 population at risk)'].mean().idxmax()
highest_incidence_year
```

#### 5.What is the correlation between the use of insecticide-treated bed nets and malaria incidence?

```python
# Convert the 'Use of insecticide-treated bed nets' column to floats, handling non-numeric values
clean_df['Use of insecticide-treated bed nets (% of under-5 population)'] = \
pd.to_numeric(clean_df['Use of insecticide-treated bed nets (% of under-5 population)'], errors='coerce')
```

```python
correlation = clean_df['Use of insecticide-treated bed nets (% of under-5 population)'].corr(
              clean_df['Incidence of malaria (per 1,000 population at risk)'])
correlation
```

#### 6.Which African country shows the most improvement in rural sanitation services over time?

```python
# Convert the 'People using safely managed sanitation services, rural' column to floats, handling non-numeric values
clean_df['People using safely managed sanitation services, rural (% of rural population)'] = \
pd.to_numeric(clean_df['People using safely managed sanitation services, rural (% of rural population)'], errors='coerce')
```

```python
# Calculate the difference in sanitation services year-over-year for each country
clean_df['Sanitation Improvement'] = clean_df.groupby('Country Name')\
['People using safely managed sanitation services, rural (% of rural population)'].diff()

most_improved_country = clean_df.groupby('Country Name')['Sanitation Improvement'].sum().idxmax()
most_improved_country
```

#### 7.What percentage of the population is Rural vs. Urban in the top 10 countries with the most malaria cases?

```python
# Identify the top 10 countries with the most malaria cases
top_10_malaria_countries = clean_df.groupby('Country Name')['Malaria cases reported'].sum().nlargest(10).index

# Filter data for these countries and select relevant columns
rural_urban_split = clean_df[clean_df['Country Name'].isin(top_10_malaria_countries)]\
[['Country Name', 'Rural population (% of total population)', 'Urban population (% of total population)']]

rural_urban_split
```
